```python
from pathlib import Path
import zipfile
import pandas as pd
import yaml
import csv
```

```python
with open("config.yaml", "r") as f:
    config = yaml.safe_load(f)

dwca_path = Path(config["input_path"])
species = config["species"]
```

```python
species
```

```python
with zipfile.ZipFile(dwca_path, "r") as z:
        with z.open("occurrence.txt") as f:
            df = pd.read_csv(f, sep="\t", low_memory=False, quoting=csv.QUOTE_NONE, encoding="utf-8")
```

```python
df
```

```python
occ_df=pd.DataFrame({"Scientific Name":df["scientificName"],
                     "Kingdom":df["kingdom"],
                     "Class":df["class"],
                     "Latitude":df["decimalLatitude"],
                     "Longitude":df["decimalLongitude"],
                     "Coordinate Uncertainty (m)":df["coordinateUncertaintyInMeters"],
                     "Country Code":df["countryCode"],
                     "Year":df["year"],
                     "Month":df["month"],
                     "Basis of Record":df["basisOfRecord"],
                     "Occurrence Status":df["occurrenceStatus"],
                     "Issue":df["issue"]})
```

```python
occ_df
```

```python
with open("config.yaml", "r") as f:
    config = yaml.safe_load(f)
    min_year=config["filters"]["min_year"]
    coord_uncertainty=config["filters"]["max_coordinate_uncertainty"]

    basis_settings = config["filters"]["allowed_basis_of_record"]
valid_types = [key for key, is_enabled in basis_settings.items() if is_enabled]
```

```python
occ_df1=occ_df.dropna(subset=["Latitude","Longitude"])
occ_df1=occ_df1[occ_df["Year"].notna() & (occ_df["Year"]>=min_year)]
occ_df1=occ_df1[occ_df["Month"].notna()]
occ_df1=occ_df1[occ_df["Coordinate Uncertainty (m)"].isna() | (occ_df["Coordinate Uncertainty (m)"]<=coord_uncertainty)]
occ_df1 = occ_df1[occ_df["Basis of Record"].isin(valid_types)]
```

```python
occ_df1
```

```python
with open("config.yaml", "r") as f:
    config = yaml.safe_load(f)
output_path = Path(config["output_dir"])
occ_df1.to_parquet(output_path, index=False)
```

```python
n_records_clean = len(occ_df1)
n_record_unclean=len(occ_df)
year_min = occ_df1["Year"].min()
year_max = occ_df1["Year"].max()
basis_counts = occ_df1["Basis of Record"].value_counts(dropna=False)
```

```python
with open("config.yaml", "r") as f:
    config = yaml.safe_load(f)
    log_path = Path(config["log_dir"])
    with open(log_path, "w") as log:
        species_name = occ_df1["Scientific Name"].iloc[0]
        log.write(f"GBIF DATA INSPECTION LOG for {species_name}\n")
        log.write("=" * 35 + "\n\n")

        log.write(f"Total records: {n_records_clean}\n")
        log.write(f"Year range: {year_min} - {year_max}\n\n")

        log.write("Basis of Record:\n")
        for k, v in basis_counts.items():
            log.write(f"  {k}: {v}\n")

        log.write("\n")

        log.write(f"Data Points before filtering: {n_record_unclean}\n")
        log.write(f"Data Points after filtering: {n_records_clean}\n")
        log.write(f"Records removed: {n_record_unclean - n_records_clean}\n")
    
```
